```python
""" prerequisites: 

1) bruker_data_process (prepreprocess step) with 'flag_make_h5_tiff', 'flag_bruker_analog', and 'flag_bruker_stim' all set 
to true

2) napeca main_parallel for sima preprocessing with 'motion_correct' flag set as true.

"""
import csv
import os
import matplotlib.pyplot as plt
import pickle
import xml.etree.ElementTree as ET

fdir = fparam['fdir'] #r'D:\bruker_data\charles_david\pl_imageactivate_02_20201113_fictive\fictive-001'
fname = fparam['fname'] #'fictive-001'
fs = 15
event_name = 'cs_plus'

eventframe_fname = 'framenumberforevents_{}.pkl'.format(fname)
```

```python
# generate trial order via txt/csv file
def csv_trial_order(fdir, fname):
    trial_order_fname = '{}_trial_order.txt'.format(fname)
    fpath = os.path.join(fdir, trial_order_fname)
    trial_order = []
    with open(fpath, 'r') as fd:
        reader = csv.reader(fd)
        for row in reader:
            trial_order.extend(row)
    return trial_order

# generate trial order via mk pt xml
def mk_pt_xml_trial_order(fdir, fname):
    mark_pt_xml_path = os.path.join(fdir, fname + '_Cycle00001_MarkPoints.xml')
    mark_pt_xml_parse = ET.parse(mark_pt_xml_path).getroot()
    trial_order = []

    for type_tag in mark_pt_xml_parse.findall('PVMarkPointElement')[1:]:

        for group_tag in type_tag.findall('PVGalvoPointElement'):

            trial_order.append((int(str.split(group_tag.attrib['Points'])[-1]))-1)
    
    return trial_order


# reload event frame numbers
with open(os.path.join(fdir, eventframe_fname), 'rb') as handle:
        tmp_dict = pickle.load(handle)

# get trial order
trial_order = mk_pt_xml_trial_order(fdir, fname)
# rewards come 2 seconds after cs plus; sort events that occur longer than 2 seconds after an event into cs event 
tmp_dict['cs_all'] = [tmp_dict['rewards'][0]] + [tmp_dict['rewards'][idx+1] for (idx,val) in enumerate(np.diff(tmp_dict['rewards'])) if val > 31  ]
tmp_dict['rewards'] = [ tmp_dict['rewards'][idx+1] for (idx,val) in enumerate(np.diff(tmp_dict['rewards'])) if val < 31  ]
tmp_dict['cs_plus'] = []
tmp_dict['cs_minus'] = []
# zipping accounts for inconsistency with planned trial numbers (trial_order) and actual ttl trial numbers (tmp_dict); gets lowest
for trial_id, cs_trial_frame in zip(trial_order, tmp_dict['cs_all']):
    # generate trial order with function
    if int(trial_id) == 0: 
        tmp_dict['cs_minus'].append(cs_trial_frame)
    else:
        tmp_dict['cs_plus'].append(cs_trial_frame)


#with open(os.path.join(fdir, eventframe_fname), 'wb') as handle:
#     pickle.dump(tmp_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python

# # reload event frame numbers
# with open(os.path.join(fdir, eventframe_fname), 'rb') as handle:
#     event_frames = pickle.load(handle)

event_frames = tmp_dict

im = bruker_marked_pts_process.load_ca_data(fdir, fname + '_sima_mc')


# extract trial data and normalize
norm_time_avg_post_stim = np.empty([len(event_frames[event_name]), im.shape[1], im.shape[2]])
for idx, trial_frame in enumerate(event_frames[event_name]):
    pre_stim = im[int(trial_frame-fs*1.5):int(trial_frame-fs*0.3),:,:]
    post_stim = im[int(trial_frame+fs*1.1):int(trial_frame+fs*2),:,:]
    pre_stim_mean = np.squeeze(np.nanmean(pre_stim, axis=0))
    pre_stim_std = np.squeeze(np.nanstd(pre_stim, axis=0))
    norm_time_avg_post_stim[idx, :, :] = np.nanmean((post_stim-pre_stim_mean), axis=0)
    
# plot time-avged normalized activity maps
to_plot = np.nanmean(norm_time_avg_post_stim,axis=0)
clims = [np.nanmin(to_plot), np.nanmax(to_plot)]

plt.figure(figsize=(7,7))
im_handle = plt.imshow(to_plot)
im_handle.set_clim(vmin=clims[0]*0.4, vmax=clims[1]*0.6)
plt.axis('off')
plt.colorbar(fraction=0.046)
```
