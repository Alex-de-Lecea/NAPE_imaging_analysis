```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
```

```python
suite2p_output_dir = "C:\\2pData\\Vijay data\\VJ_OFCVTA_7_D8_trained\\suite2p\plane0\\"
```

```python
ops = np.load(suite2p_output_dir + "ops.npy", allow_pickle = True).item()
ops
```

```python
ops['meanImg'].shape
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,10) )
im = ax.imshow(ops['meanImg'], cmap='gray')
```

```python
mc_dims = [ops['nframes'], ops['Lyc'], ops['Lxc']]
raw_dims = [ops['nframes'], ops['Ly'], ops['Lx']]
filename = str.split( str.split(ops['h5py'],'/')[-1] , '.')[0]
filename
```

```python
suite2p_corr = np.fromfile(suite2p_output_dir + "data.bin",  dtype=np.int16).reshape(raw_dims)
suite2p_corr.shape
```

```python
mean_img_corr = np.mean( suite2p_corr, axis = 0 )
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,10) )
im = ax.imshow(mean_img_corr, cmap='gray')
im.set_clim(vmin=0, vmax=135)

#fig.colorbar(im);
```

```python
# save cut data as h5
hf = h5py.File(suite2p_output_dir + filename + '_mc.h5', 'w')
hf.create_dataset('data', data=suite2p_corr)
hf.close()
```

```python
# for suite2p with raw bin

suite2p_raw = np.fromfile(suite2p_output_dir + "data_raw.bin",  dtype=np.int16).reshape(raw_dims)
mean_img_raw = np.mean( suite2p_raw, axis = 0 )
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20,20) )
im = ax[0].imshow(mean_img_raw, cmap='gray')
ax[0].set_title('Raw',fontsize=20)
im.set_clim(vmin=0, vmax=110)
im1 = ax[1].imshow(mean_img_corr, cmap='gray')
ax[1].set_title('MoCorr',fontsize=20)
im1.set_clim(vmin=0, vmax=110)
```
