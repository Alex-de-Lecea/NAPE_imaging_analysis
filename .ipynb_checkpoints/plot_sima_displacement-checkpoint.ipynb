```python
import numpy as np
import os
import pickle

import matplotlib.pyplot as plt 
```

```python
#filename = 'VJ_OFC_7_240_D6'
filename = 'itp_lhganiii_bl3_935_0001'

#folder = 'C:/2pData/Vijay data/VJ_OFC_7_D9_trained/'
folder = 'C:/2pData/Ivan/itp_lhganiii_bl3_678/'
```

```python
displacement_file = open( os.path.join(folder, filename + '_mc.sima/displacement.pkl'), "rb" )
print(displacement_file)
displacement = pickle.load(displacement_file)
```

```python
print(displacement[0].shape)

disp_np = np.squeeze( np.array(displacement[0]) )
```

```python
num_frames, y_lines, x_pixels  = disp_np.shape
print(disp_np.shape)
disp_meanx = np.mean( disp_np, axis=1 )
```

```python
fps = 30
num_frames = 300
end_time = 100/fps
tvec = np.linspace(0,end_time,num=num_frames)
```

```python
plt.plot(tvec, disp_meanx[:num_frames,0])
plt.plot(tvec, disp_meanx[:num_frames,1])
plt.xlabel('Time [s]')
plt.ylabel('Mean Disp')
```
